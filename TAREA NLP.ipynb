```python
import spacy
nlp = spacy.load("es_core_news_sm")
def normalize(text):
    doc = nlp(text)
    words = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
    lexical_tokens = [t.lower() for t in words if len(t) > 3 and     
    t.isalpha()]
    return lexical_tokens
```

# Lemmatization

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
docx = nlp(u"Prebiotic chemistry can be understood as the chemistry which occurred before life began or the chemistry which led to life on Earth, and possibly on other planets. The origin of life remains unexplained despite decades (or perhaps centuries, depending on where one historically marks the starting point) of research. A considerable amount of study has provided compelling details as to how it might have occurred and whether it is likely to be a universal phenomenon. The development of modern thought on the topic has a long and winding history and has been modified to adapt to developments in other fields, including astronomy, biology, chemistry, and geology. This review covers the synthesis of small organic compounds; however, it should be borne in mind that the major uncertainties revolve around how these compounds self-organize into self-replicating systems. It’s important to understand the equilibrium between the chemical compound in the atmosphere and then, we can stablish a cycle in the inside of mars.")
```

```python
for word in docx:
    print(word.text,word.lemma_,word.pos_)
```

```python
for token in docx:
    print(token.text, '\t', token.pos_, '\t', token.lemma_)
```

```python
def show_lemmas(text):
    for token in text:
        print(f'{token.text:{12}} {token.pos_:{6}}  {token.lemma_}')
```

```python
docy = nlp(u"Prebiotic chemistry can be understood as the chemistry which occurred before life began or the chemistry which led to life on Earth, and possibly on other planets. The origin of life remains unexplained despite decades (or perhaps centuries, depending on where one historically marks the starting point) of research. A considerable amount of study has provided compelling details as to how it might have occurred and whether it is likely to be a universal phenomenon. The development of modern thought on the topic has a long and winding history and has been modified to adapt to developments in other fields, including astronomy, biology, chemistry, and geology. This review covers the synthesis of small organic compounds; however, it should be borne in mind that the major uncertainties revolve around how these compounds self-organize into self-replicating systems. It’s important to understand the equilibrium between the chemical compound in the atmosphere and then, we can stablish a cycle in the inside of mars.")

show_lemmas(docy)
```
